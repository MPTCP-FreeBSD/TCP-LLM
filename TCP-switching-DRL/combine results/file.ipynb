```python
import dpkt
import socket
from collections import defaultdict
import pandas as pd
import matplotlib.pyplot as plt

def calculate_rates(pcap_file):
    # Read the pcap file
    with open(pcap_file, 'rb') as f:
        pcap = dpkt.pcap.Reader(f)

        # Initialize counters and storage
        packet_counts = defaultdict(int)
        packet_sizes = defaultdict(int)
        seq_numbers = defaultdict(set)
        ack_numbers = defaultdict(set)
        lost_packets = defaultdict(int)
        total_packets = defaultdict(int)
        intervals = defaultdict(lambda: defaultdict(list))

        # Process each packet
        for timestamp, buf in pcap:
            eth = dpkt.ethernet.Ethernet(buf)
            if isinstance(eth.data, dpkt.ip.IP):
                ip = eth.data
                if isinstance(ip.data, dpkt.tcp.TCP):
                    tcp = ip.data
                    src_ip = socket.inet_ntoa(ip.src)
                    dst_ip = socket.inet_ntoa(ip.dst)
                    size = len(buf)
                    time_sec = int(timestamp)

                    if (src_ip, dst_ip) not in intervals:
                        intervals[(src_ip, dst_ip)] = defaultdict(list)

                    # Track packet count and sizes
                    packet_counts[(src_ip, dst_ip)] += 1
                    packet_sizes[(src_ip, dst_ip)] += size

                    # Track sequence and acknowledgment numbers
                    seq_numbers[(src_ip, dst_ip)].add(tcp.seq)
                    ack_numbers[(src_ip, dst_ip)].add(tcp.ack)

                    # Track packets per second
                    intervals[(src_ip, dst_ip)][time_sec].append(size)

        # Calculate sending rate and loss rate
        sending_rates = {}
        loss_rates = {}

        for key in intervals:
            src_ip, dst_ip = key
            times = sorted(intervals[key].keys())
            if len(times) > 1:
                sending_rate = []
                total_sent = 0

                for i in range(len(times) - 1):
                    interval_time = times[i + 1] - times[i]
                    if interval_time > 0:
                        sent_in_interval = sum(intervals[key][times[i]])
                        sending_rate.append(sent_in_interval / interval_time)
                        total_sent += sent_in_interval

                total_lost = len(seq_numbers[key]) - len(ack_numbers[key])
                total_packets_sent = len(seq_numbers[key])
                loss_rate = total_lost / total_packets_sent if total_packets_sent > 0 else 0

                sending_rates[key] = sum(sending_rate) / len(sending_rate) if len(sending_rate) > 0 else 0
                loss_rates[key] = loss_rate

        return sending_rates, loss_rates

def main():
    # Path to the pcap file
    pcap_file = '/Users/shyamshrestha/Downloads/modified2.pcap'  # Replace with the path to your pcap file

    # Calculate rates
    sending_rates, loss_rates = calculate_rates(pcap_file)

    # Print results
    for key in sending_rates:
        src_ip, dst_ip = key
        print(f"Connection from {src_ip} to {dst_ip}:")
        print(f"  Sending Rate: {sending_rates[key]} bytes/second")
        print(f"  Loss Rate: {loss_rates[key] * 100}%")

if __name__ == "__main__":
    main()
```
